```python
%matplotlib inline

import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import norm
import scipy
from pdb import set_trace as st
import sympy as sy

theta1 = [0, 1]
theta2 = [1, 2]
x = np.linspace(-8, 8, 100)
# x1 and x2 denote p(x|w_1) and p(x|w_2)
x1 = norm.pdf(x, theta1[0], theta1[1])  
x2 = norm.pdf(x, theta2[0], theta2[1]) 


fig, ax = plt.subplots(figsize=(8,6))
plt.grid(which='both',axis='both', color='grey',linestyle=':')
plt.plot(x, x1,label='p(x|W1)', color='tab:blue', linewidth=3)
plt.plot(x, x2,label='p(x|W2)', color='tab:green', linewidth=3)

idx = np.argwhere(np.diff(np.sign(x2 - x1)))

#print(x[idx][0])
#print(x[idx][1])

plt.axvline(x[idx][0], 0, 1, label= 'Class 1 if  p(W1|x) > p(W2|x)',color = 'red')
plt.axvline(1.15, 0, 1, color = 'red')







plt.title('Class-conditional Density (Likelihood)')
plt.xlabel("X")
plt.ylabel("Likelihood")

plt.legend(fancybox=True, framealpha=0.5, fontsize=18)
plt.show()
```

```python
evidence = (.5 * x1) + (0.5 *x2)



px1= (0.5 *x1) / evidence 

px2 = (0.5 * x2) / evidence
  






fig, ax = plt.subplots(figsize=(8,6))

plt.grid(which='both',axis='both', color='grey',linestyle=':')
plt.plot(x, px1 , label='p(W1|x)', color='tab:blue', linewidth=3)
plt.plot(x, px2, label='p(W2|x)', color='tab:green', linewidth=3)

plt.axvline(x[idx][0], 0, 1, label= 'Class 1 if  p(W1|x) > p(W2|x)',color = 'red')
plt.axvline(1.15, 0, 1, color = 'red')






plt.title('Posterior distribution')
plt.xlabel("X")
plt.ylabel("Probability")

plt.legend(fancybox=True, framealpha=0.5, fontsize=10)
plt.show()
```

```python
%matplotlib inline

import numpy as np
from math import e
import matplotlib.pyplot as plt
from scipy.stats import norm
from pdb import set_trace as st


x = np.linspace(-0, 5, 100)
theta = np.linspace(0,5,100)

px1 = e**-x
px2 = theta * e**(theta * -2) 
 

fig, ax = plt.subplots(figsize=(8,6))
plt.grid(which='both',axis='both', color='grey',linestyle=':')
plt.plot(x, px1, label='p(x|theta) versus x', color='tab:blue', linewidth=3)
plt.plot(theta,px2, label = 'p(x|theta) versus theta')

plt.title('Class-conditional Density (Likelihood)')
plt.legend(fancybox=True, framealpha=0.5, fontsize=18)
plt.show()
```
